```python
import matplotlib.pyplot as plt

from chapters.work_with_data import plot_histogram
```

```python
import random
from chapters.probability import inverse_normal_cdf

random.seed(0)

# Равномерное распределение между -100 и 100
uniform = [200 * random.random() - 100 for _ in range(10000)]

# Нормальное распределение со средним 0, стандартным отклонением 57
normal = [57 * inverse_normal_cdf(random.random()) for _ in range(10000)]


plot_histogram(uniform, 10, "Равномерное распределение")
plot_histogram(normal, 10, "Нормальное распределение")
```

```python
# Двумерные данные
def random_normal():
    """Возвращает случайную выборку из стандартного
       нормального распределения"""
    return inverse_normal_cdf(random.random())

xs = [random_normal() for _ in range(1000)]
ys1 = [x + random_normal() / 2 for x in xs]
ys2 = [-x + random_normal() / 2 for x in xs]
```

```python
from chapters.statistics import correlation

plt.scatter(xs, ys1, marker='.', color='black', label='ys1')
plt.scatter(xs, ys2, marker='.', color='gray', label='ys2')
plt.xlabel('xs')
plt.ylabel('ys')
plt.legend(loc=9)
plt.title("Совсем разным совместные распределения")
plt.show()
print(correlation(xs, ys1))
print(correlation(xs, ys2))
```

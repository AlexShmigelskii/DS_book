```python
import matplotlib.pyplot as plt
import math

def B(alpha: float, beta: float) -> float:
    """Нормализирующая константа, чтобы полная
       вероятность в суме составляла 1"""
    return math.gamma(alpha) * math.gamma(beta) / math.gamma(alpha + beta)


def beta_pdf(x: float, alpha: float, beta: float) -> float:
    if x <= 0 or x >= 1:  # за пределами [0, 1] нет веса
        return 0
    return x ** (alpha - 1) * (1 - x) ** (beta - 1) / B(alpha, beta)
```

```python
xs = [x / 100.0 for x in range(100 + 1)]

plt.plot(xs, [beta_pdf(x, alpha=1, beta=1) for x in xs], '-', label='alpha=1, beta=1')
plt.plot(xs, [beta_pdf(x, alpha=10, beta=1) for x in xs], '--', label='alpha=10, beta=1')
plt.plot(xs, [beta_pdf(x, alpha=8, beta=20) for x in xs], '-.', label='alpha=8, beta=20')
plt.plot(xs, [beta_pdf(x, alpha=30, beta=30) for x in xs], '.', label='alpha=30, beta=30')


plt.axis([0, 1, 0, 10])
plt.legend()
plt.title("Примеры бета-распределений")
plt.show()
```
